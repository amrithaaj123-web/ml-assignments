```python
def group_anagrams(words):
    d = {}
    for w in words:
        k = ''.join(sorted(w))
        if k in d:
            d[k].append(w)
        else:
            d[k] = [w]
    return list(d.values())
print(group_anagrams(["eat","tea","tan","ate","nat","bat"]))
```

```python
def flood_fill(img, r, c, new):
    old = img[r][c]
    
    if old == new:
        return img
    
    def fill(r, c):
        if r<0 or c<0 or r>=len(img) or c>=len(img[0]) or img[r][c] != old:
            return
        
        img[r][c] = new
        
        fill(r+1, c)
        fill(r-1, c)
        fill(r, c+1)
        fill(r, c-1)
    
    fill(r, c)
    return img
```

```python
def setZeroes(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    row_zero = [False] * rows
    col_zero = [False] * cols
    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] == 0:
                row_zero[i] = True
                col_zero[j] = True
    for i in range(rows):
        for j in range(cols):
            if row_zero[i] or col_zero[j]:
                matrix[i][j] = 0
                return matrix
matrix = [[0,2,3],
          [4,5,6],
          [7,8,9]]
print(setZeroes(matrix))
print(setZeroes(matrix))
```

```python
def maxAverage(nums, k):
    window_sum = sum(nums[:k])
    max_sum = window_sum
    for i in range(k, len(nums)):
        window_sum = window_sum + nums[i] - nums[i - k]
        if window_sum > max_sum:
            max_sum = window_sum
    return max_sum / k
nums = [1, 2, 3, 4, 5]
k = 2
print(maxAverage(nums, k))
```
